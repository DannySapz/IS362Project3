```python
#import pandas, numpy, and sqlite3 
import pandas as pd
import numpy as np
import sqlite3
```

```python
#establish a connection to the sqlite file
connection = sqlite3.connect("C://Users//danie//Downloads//Chinook Database//Chinook_Sqlite.sqlite")
#get the last names and first names of the customers in the customers table
#customer Id is useful since it can be used as a key between Invoice Line and Customer Table to match up information
customer = pd.DataFrame(pd.io.sql.read_sql("SELECT LastName, FirstName, CustomerId FROM Customer", connection))
print customer
```

```python
#get the invoice id and customer id from the invoice table
invoice = pd.DataFrame(pd.io.sql.read_sql("SELECT InvoiceId, CustomerId FROM Invoice", connection))
#use the customer id as a key to merge customer and invoice table columns
df3 = pd.merge(customer, invoice, on=['CustomerId'])
print df3
```

```python
#get the invoice id and track id from the invoiceline table
invoice_line = pd.DataFrame(pd.io.sql.read_sql("SELECT InvoiceId, TrackId FROM InvoiceLine", connection))
#use the invoice id as a key to connect the invoice line table to the previous dataframe
df4 = pd.merge(df3, invoice_line, on=['InvoiceId'])
print df4
```

```python
#get the name, albumid, and trackid from the track table
track_title = pd.DataFrame(pd.io.sql.read_sql("SELECT Name, AlbumId, TrackId FROM Track", connection))
#use the track id as a key to establish a connection between the previous dataframe and track_title
df5 = pd.merge(df4, track_title, on=['TrackId'])
print df5
```

```python
#get the titile and album id from the album table
album_title = pd.DataFrame(pd.io.sql.read_sql("SELECT Title, AlbumId FROM Album", connection))
#use album id as the key to establish a connection between the previous dataframe with album_title
df6 = pd.merge(df5, album_title, on=['AlbumId'])
print df6
```

```python
#make a copy of the dataframe
df7 = df6.copy()
print df7
```

```python
#list of columns
df7.columns
```

```python
# delete the unnecessary columns for the final result
del df7['CustomerId']
del df7['InvoiceId']
del df7['TrackId']
del df7['AlbumId']
print df7
```

```python
# sort the last names by ascending values
result = df7.sort_values(by='LastName',ascending=True)
print result
```

```python
#reset the index since it was messed up from all the sorting and dropping of columns
df = result.reset_index(drop=True)
print df
```
